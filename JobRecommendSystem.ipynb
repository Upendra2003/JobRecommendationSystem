### Seperating 200 rows of each job role and exporting the dataframe

```python
new_df = pd.DataFrame(columns=df.columns)
```

```python
desired_rows=90

job_roles=['SDE', 'Android', 'Frontend', 'Java', 'Testing', 'IOS', 'Backend', 'Web', 'Python', 'Database', 'Mobile']

for job_role in job_roles:
    job_role_rows = df[df['Job Roles'] == job_role]
    for i in range(0,desired_rows):
        new_df=new_df.append(job_role_rows.iloc[i],ignore_index=True)
```

we have to find the domain of each company which can be accesed in the url in the company url

```python
import requests

def find_company_domain(url):
    try:
        response=requests.get(url,timeout=10)
        response.raise_for_status()
        data=response.json()
        if data:
            first_item=data[0]
            print('Completed domain :',first_item['name'])
            return first_item['domain']
        else:
            return 'nan'
    except requests.exceptions.RequestException as e:
        print(f"Error:{e}")
```
