```python
s_df=pd.read_csv('Reduced_salary_list.csv')
s_df.head()
```

```python
category_keywords_mapping = {
    'SDE': {'Keywords': ['Software Development Engineer', 'Software Engineer', 'Developer'], 'Tech Stacks': ['Java', 'C++', 'Python']},
    'Android': {'Keywords': ['Android', 'Mobile', 'Mobile Developer'], 'Tech Stacks': ['Java', 'Kotlin', 'Android Studio']},
    'Frontend': {'Keywords': ['Frontend', 'UI', 'UX', 'Web Developer'], 'Tech Stacks': ['HTML', 'CSS', 'JavaScript', 'React', 'Vue', 'Angular']},
    'Java': {'Keywords': ['Java', 'Java Developer'], 'Tech Stacks': ['Java', 'Spring', 'Hibernate']},
    'Testing': {'Keywords': ['Testing', 'Test Engineer'], 'Tech Stacks': ['Selenium', 'JUnit', 'TestNG']},
    'IOS': {'Keywords': ['iOS', 'Mobile', 'iOS Developer'], 'Tech Stacks': ['Swift', 'Objective-C', 'Xcode']},
    'Backend': {'Keywords': ['Backend', 'Server', 'Backend Developer'], 'Tech Stacks': ['Node.js', 'Django', 'Flask', 'Express']},
    'Web': {'Keywords': ['Web', 'Web Developer'], 'Tech Stacks': ['HTML', 'CSS', 'JavaScript', 'React', 'Vue', 'Angular']},
    'Python': {'Keywords': ['Python', 'Python Developer'], 'Tech Stacks': ['Python', 'Django', 'Flask']},
    'Database': {'Keywords': ['Database', 'DBA', 'Database Administrator'], 'Tech Stacks': ['SQL', 'MySQL', 'MongoDB']},
    'Mobile': {'Keywords': ['Mobile', 'Mobile Developer'], 'Tech Stacks': ['Java', 'Kotlin', 'Swift', 'Objective-C']},
}

def get_info(job_title):
    for category, info in category_keywords_mapping.items():
        if any(category.lower() in job_title.lower() for keyword in info['Keywords']):
            return {'Keywords': info['Keywords'], 'Tech Stacks': info['Tech Stacks']}
    return {'Keywords': [], 'Tech Stacks': []}

s_df[['Keywords', 'Tech Stacks']] = s_df['Job Title'].apply(lambda x: pd.Series(get_info(x)))   
```

```python
s_df[s_df['Job Roles']=='IOS'].head()
```

```python
s_df['Job Title']=s_df['Job Title'].apply(lambda x:x.split())
s_df['Location']=s_df['Location'].apply(lambda x:x.split())
s_df['Employment Status']=s_df['Employment Status'].apply(lambda x:x.split())
```

```python
s_df.head()
```

```python
s_df['tags']=s_df['Job Title']+s_df['Location']+s_df['Employment Status']+s_df['Job Roles']+s_df['Keywords']+s_df['Tech Stacks']
```

```python
new_df=s_df[['Company Name','Salary','tags','Job Roles']]
```
