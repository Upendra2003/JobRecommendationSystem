# Adding Logo column

```python
import requests 

def find_company_logo(url):
    try:
        response=requests.get(url,timeout=10)
        response.raise_for_status()
        data=response.json()
        if data:
            first_item=data[0]
            print('Completed logo :',first_item['logo'])
            return first_item['logo']
        else:
            return 'nan'
    except requests.exceptions.RequestException as e:
        print(f"Error:{e}")
```
